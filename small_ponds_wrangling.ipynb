## Now let's look at the graph

The number of connections each person, address, etc has is one thing. But it's more interesting to look at how those connections are connected to one another. Let's see if we can figure out who is spinning a web, and who's just well-connected. 

This takes a little setup. Feel free to skip ahead.

```python
# use these to build graph and draw it
# scsc_merged = pd.merge(scs_merged, companies_df, left_on = 'company_id', right_on = 'id')
# sasa_merged = pd.merge(sas_merged, addresses_df, left_on = 'address_id', right_on = 'id')

import graph_tool as gt
from graph_tool.all import *
```

```python
g = Graph(directed=False)
object_ids = g.new_vertex_property("string")
object_types = g.new_vertex_property("string")
object_labels = g.new_vertex_property("string")

g.vertex_properties["object_ids"] = object_ids
g.vertex_properties["object_types"] = object_types
g.vertex_properties["object_labels"] = object_labels

company_node_hash = {}
stakeholder_node_hash = {}
address_node_hash = {}
```

```python
for row in sasa_merged[(sasa_merged['degree'] >= 94)].iterrows():
    address_id = row[1]['calc_id']
    stakeholder_id = row[1]['stakeholder_id']
    
    if stakeholder_id not in stakeholder_node_hash:
        vs = g.add_vertex()
        stakeholder_node_hash[stakeholder_id] = vs
        object_ids[vs] = str(stakeholder_id)
        object_types[vs] = row[1]['type']
        object_labels[vs] = row[1]['name']
    else:
        vs = stakeholder_node_hash[stakeholder_id]

    if address_id not in address_node_hash:
        va = g.add_vertex()
        address_node_hash[address_id] = va
        object_ids[va] = str(address_id)
        object_types[va] = 'address'
        object_labels[va] = str(row[1]['street']) + ' ' + str(row[1]['city']) + ' ' + str(row[1]['postal']) 
    else:
        va = address_node_hash[address_id]

    g.add_edge(vs,va)

print "Saving the graph..."
g.save("stakeholder-address-graph.xml.gz")

print "Total nodes"
print g.num_vertices()

print "Total edges"
print g.num_edges()
```

```python
g = load_graph("stakeholder-address-graph.xml.gz")
#bad_v = g.vertex(90)
#g.remove_vertex(bad_v)
pos = sfdp_layout(g,max_iter=30)
graph_draw(g, pos=pos, output_size=(800, 800), inline=True)
```

## What's so special about that cluster. 

Are those entities more connected than everyone else?

**Nope**. Here we're sizing the dots based on their degree. Turns out the most connected entities are spread pretty evenly around the network. Meh.

```python
deg = g.degree_property_map("out")
deg.a = 2 * (sqrt(deg.a) * 0.5 + 0.1)
#ebet = betweenness(g)[0]
pos = sfdp_layout(g,max_iter=30)
graph_draw(g, pos=pos, output_size=(800, 800), vertex_size=deg, inline=True)
```

## The "closeness" finds big fish in small ponds

So a person or address that is central (really important) to a distinct cluster should pop out as an exceptionally large dot. 

**Bummer**. Everyone's closeness is pretty similar.

```python
eclose = closeness(g)
g.vertex_properties["betweenness_value"] = ebet
eclose.a /= eclose.a.max() / 20.
pos = sfdp_layout(g,max_iter=30)
graph_draw(g, pos=pos, output_size=(800, 800), vertex_size=eclose, inline=True)
```

## ...and bingo.

**Betweenness** finds "hubs" in a network. Entities through which information must flow. Frankly this doesn't really matter, what's important is that this is the first measure of relative importance in the network where a handful of entities really stick out. There should be thousands of dots in the graph, but all except a small handful are too small to see. 

**So who are these entities with exceptionally high *betweenness*?**

```python
ebet = betweenness(g)[0]
g.vertex_properties["betweenness_value"] = ebet
ebet.a /= ebet.a.max() / 40.
pos = sfdp_layout(g,max_iter=30)
graph_draw(g, pos=pos, output_size=(800, 800), vertex_size=ebet, inline=True)
```

```python
rows_list = []
for v in g.vertices():
    obj = {
        'label': object_labels[v],
        'entity_id': object_ids[v],
        'entity_type': object_types[v],
        'betweenness': ebet[v]
    }
        
    rows_list.append(obj)

centrality_df = pd.DataFrame(rows_list)
```
